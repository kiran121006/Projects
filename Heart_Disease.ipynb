## **3. Dataset** **Overview**

*   The dataset contains a total of 303 patient records.
*  It includes 14 attributes representing various clinical and physiological parameters.

1. age	: age in years.
2. sex: sex (1 = male; 0 = female).
3. cp: chest pain type.

   *    Value 1: typical angina  (Heart-related chest pain during exertion)
   *    Value 2: atypical angina  (Unclear or unusual chest pain)
   *    Value 3: non-anginal pain  (Chest pain not related to heart)
   *    Value 4: asymptomatic  (No chest pain but possible heart disease)
4. trestbps: resting blood pressure (in mm Hg on admission to the hospital).
5. chol: serum cholesterol in mg/dl.
6. fbs: (fasting blood sugar > 120 mg/dl)  (1 = true; 0 = false).
7. restecg: resting electrocardiographic results
     *   Value 0: normal
     *  Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)
     * Value 2: showing probable or definite left ventricular hypertrophy by Estes' criteria.
8. thalach: maximum heart rate achieved.
9. exang: exercise induced angina (1 = yes; 0 = no).
10. oldpeak = ST depression induced by exercise relative to rest.(how much the heart shows abnormal stress during exercise)
11. slope: the slope of the peak exercise ST segment
    *   Value 1: upsloping
    *    Value 2: flat
    *    Value 3: downsloping.

12. ca: no. of major vessels colored by fluoroscopy
13. thal: Thalassemia type  (type of blood flow defect)

       3 = normal; 6 = fixed defect; 7 = reversable defect.
14. target:
  - 0 → No heart disease (Normal condition)  
  - 1 → Mild heart disease  
  - 2 → Moderate heart disease  
  - 3 → Severe heart disease  
  - 4 → Very severe heart disease

## **Cholesterol Classification**


Cholesterol levels are categorized as:
- Normal: < 200
- Borderline: 200–240
- High: > 240


```python
def chol_category(x):
    if x < 200:
        return 'Normal'
    elif x <= 240:
        return 'Borderline'
    else:
        return 'High'

df['chol_level'] = df['chol'].apply(chol_category)
df['chol_level'].value_counts()
```

```python
#  Cholesterol Level Distribution
df['chol_level'] = pd.cut(df['chol'],
                          bins=[0,200,240,600],
                          labels=['Normal', 'Borderline High', 'High'])
plt.figure(figsize=(6,6))

df['chol_level'].value_counts().plot.pie(
    autopct='%1.1f%%',
    startangle=90
)

plt.title("Cholesterol Level Distribution")
plt.ylabel("")  # removes y-axis label
plt.show()
```

## **Insight:**

A significant number of patients have elevated cholesterol levels, indicating that high cholesterol is common.

```python
# Cholesterol Level vs Heart Disease
df['target_label'] = df['target'].replace({0: 'No Disease', 1: 'Disease', 2: 'Disease', 3: 'Disease', 4: 'Disease'})

plt.figure(figsize=(7,4))
sns.countplot(x='chol_level', hue='target_label', data=df)

plt.title("Cholesterol Level vs Heart Disease")
plt.xlabel("Cholesterol Category")
plt.ylabel("Number of Patients")

plt.show()
```



## **Insight:**

# ***6. Correlation Analysis***

Correlation analysis helps identify relationships between variables.

Which features are most related to heart disease?

```python
#Correlation Matrix

# select only numeric columns
df_numeric = df.select_dtypes(include=['number'])

plt.figure(figsize=(12,8))

sns.heatmap(df_numeric.corr(),
            annot=True,
            cmap='coolwarm',
            fmt=".2f",
            linewidths=0.5)

plt.title("Correlation Matrix of Heart Disease Dataset")
plt.show()
```

## **Insight:**


The heatmap shows that cp, thalach, and exang have stronger relationships with heart disease, while chol and trestbps have weaker correlations.
This suggests that heart disease is affected by a combination of multiple features.

```python
# Correlation with Target
df_numeric.corr()['target'].sort_values().plot(kind='barh')
plt.title("Correlation with Target")
plt.show()
```

```python
# Strongly Correlated Features with Heart Disease
corr = df_numeric.corr()['target']

strong_corr = corr[abs(corr) > 0.3]
print(strong_corr)
```

## **7. GroupBy Analysis**

```python
# Average Feature Values by Target Variable
df.groupby('target')[['age','trestbps','chol','thalach']].mean().round(2)
```

## **GroupBy Analysis Insights:**

The grouped data shows clear differences between patients with and without heart disease.

- Patients with heart disease tend to have a slightly higher average age, indicating that risk increases with age.

- Cholesterol levels (chol) do not show a very strong difference, indicating that it may not be the most dominant factor alone.
- Maximum heart rate achieved (thalach) is lower in heart disease patients, suggesting reduced cardiovascular efficiency.

Overall, this analysis indicates that heart rate and blood pressure are stronger indicators compared to cholesterol.
