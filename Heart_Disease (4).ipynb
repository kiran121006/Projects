## **2. Dataset** **Overview**

*   The dataset contains a total of 303 patient records.
*  It includes 14 attributes representing various clinical and physiological parameters.

1. age	: age in years.
2. sex: sex (1 = male; 0 = female).
3. cp: chest pain type.

   *    Value 1: typical angina
   *    Value 2: atypical angina
   *    Value 3: non-anginal pain
   *    Value 4: asymptomatic
4. trestbps: resting blood pressure (in mm Hg on admission to the hospital).
5. chol: serum cholestoral in mg/dl.
6. fbs: (fasting blood sugar > 120 mg/dl)  (1 = true; 0 = false).
7. restecg: resting electrocardiographic results
     *   Value 0: normal
     *  Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)
     * Value 2: showing probable or definite left ventricular hypertrophy by Estes' criteria.
8. thalach: maximum heart rate achieved.
9. exang: exercise induced angina (1 = yes; 0 = no).
10. oldpeak = ST depression induced by exercise relative to rest.
11. slope: the slope of the peak exercise ST segment
    *   Value 1: upsloping
    *    Value 2: flat
    *    Value 3: downsloping.

12. ca: no. of major vessels colored by fluoroscopy
13. thal: Thalassemia type

       3 = normal; 6 = fixed defect; 7 = reversable defect.
14. target:
  - 0 → No heart disease (Normal condition)  
  - 1 → Mild heart disease  
  - 2 → Moderate heart disease  
  - 3 → Severe heart disease  
  - 4 → Very severe heart disease

## ***Cholesterol Classification***


Cholesterol levels are categorized as:
- Normal: < 200
- Borderline: 200–240
- High: > 240


```python
def chol_category(x):
    if x < 200:
        return 'Normal'
    elif x <= 240:
        return 'Borderline'
    else:
        return 'High'

df['chol_level'] = df['chol'].apply(chol_category)
df['chol_level'].value_counts()
```

```python
plt.figure(figsize=(7,4))
sns.countplot(x='chol_level', data=df)
plt.title("Cholesterol Level Distribution")
plt.show()
```

# ***Insight:***

A significant number of patients have elevated cholesterol levels, indicating that high cholesterol is common.

```python
df['target_label'] = df['target'].replace({0: 'No Disease', 1: 'Disease', 2: 'Disease', 3: 'Disease', 4: 'Disease'})

plt.figure(figsize=(7,4))
sns.countplot(x='chol_level', hue='target_label', data=df)

plt.title("Cholesterol Level vs Heart Disease")
plt.xlabel("Cholesterol Category")
plt.ylabel("Number of Patients")

plt.show()
```

# ***Insight:***

Higher cholesterol levels are associated with an increased occurrence of heart disease.

***9. Correlation Heatmap:***

Which features are most related to heart disease?

```python
import seaborn as sns
import matplotlib.pyplot as plt

# select only numeric columns
df_numeric = df.select_dtypes(include=['number'])

plt.figure(figsize=(12,6))
sns.heatmap(df_numeric.corr(), annot=True, cmap='coolwarm')

plt.title("Correlation Matrix of Heart Disease Dataset")
plt.show()
```

# ***Correlation Analysis***


Correlation analysis helps identify relationships between variables.

It is observed that features like chest pain type, maximum heart rate,
and exercise-induced angina show stronger correlation with heart disease,
making them important predictors.


```python
df_numeric = df.select_dtypes(include=['number'])

df_numeric.corr()['target'].sort_values(ascending=False)
```

```python
corr = df_numeric.corr()['target']

strong_corr = corr[abs(corr) > 0.3]
print(strong_corr)
```
